```python
img2 = cv2.imread("Dog_Breeds.jpg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
fig,sub=plt.subplots(figsize=(5,5))
sub.imshow(img2)
sub.axis("off")
plt.show()
```

```python
img2.shape
```

```python
img2_resized=cv2.resize(img2,None, fx=0.10,fy=0.10)
fig,sub=plt.subplots(figsize=(5,5))
sub.imshow(img2_resized)
sub.axis("off")
plt.show()
```

```python
img2_resized.shape
```

```python
kernel_sharping=np.array(
    [[0, -1, 0],
    [-1, 9, -1],
     [0, -1, 0]])

sharpen_img=cv2.filter2D(img2,-1,kernel_sharping)
# sharpen_img_rgb= cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
fig,sub=plt.subplots(figsize=(5,5))
sub.imshow(sharpen_img)
sub.axis("off")
plt.show()
```

```python
#blurred image 

blurred_img = cv2.GaussianBlur(img2, (25, 25), 0)
fig,sub=plt.subplots(figsize=(5,5))
sub.imshow(blurred_img)
sub.axis("off")
plt.show()
```

```python
img2 = cv2.imread("Dog_Breeds.jpg")
img2= cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

_,thresh = cv2.threshold(img2, 127, 255, cv2.THRESH_BINARY)

fig, sub = plt.subplots(figsize=(5,5))
sub.imshow(thresh, cmap="gray")
sub.axis("off")
plt.show()
```
